```python
file='/home/echo/environments/旅游城市.txt'
f=open(file,'r')
ff=f.readlines()
a = []
for line in ff:
    line=line.rstrip("\n")
    a.append(line)
c =''
for b in a:
    if(b != ''):
        c = c +b
d = c.split(' ')
data_b = []
for e in d:
    if(e != ''):
        data_b.append(e)
```
