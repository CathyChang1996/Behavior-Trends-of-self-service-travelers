```python
url2='https://www.mafengwo.cn/wenda/area-187613.html'
import time
from selenium import webdriver
browser = webdriver.Chrome()

browser.get(url2)
globe=[]
for i in range(50):
    time.sleep(2)
    try:
        all_titles=[]
        titles=browser.find_elements_by_css_selector('div.title')
        for t in titles:
            title=t.text
            all_titles.append(title)
        globe.extend(all_titles)
        browser.execute_script('window.scrollTo(0, document.body.scrollHeight/1.2);')#test several numbers to choose a suitable one
        next_page = browser.find_element_by_partial_link_text('加载更多')
        next_page.click()
    except Exception as e:
        print(e)
        print('Error on page %s' % i)
```

```python
titles = "".join(all_titles)
```

```python
f = open("titles.txt",'w') 
f.write(titles)
f.close()   
```

```python
import jieba
import re
```

```python
with open('titles.txt',encoding='utf-8',errors='ignore') as f:
    word = f.read()
    resultword = re.sub("[A-Za-z0-9\[\“\”\：\（\）\——\？\、\，\?\`\~\!\@\#\$\^\&\*\(\)\=\|\{\}\'\:\;\'\,\[\]\.\<\>\/\?\~\！\@\#\\\&\*\%]","",word)
    wordlist_after_jieba = jieba.cut(resultword)
    wl_space_split = " ".join(wordlist_after_jieba)
    wordlist = wl_space_split.split()
```

```python
word_count= []
```

```python
wordCount = {}
for word in wordlist:
    if word in wordCount:
        wordCount[word] = wordCount[word]+1
    else:
        wordCount[word] = 1
        
with open('stopwords.txt',encoding='utf-8',errors='ignore') as f:
    stopwords = f.read().split()
    stopwords.extend(['到','这','一','最','与','吧','好','但是','的话','回复','就是','还是','因为','一定','如果','很多','比较','个','人','这样','这么','从','可以','哪些','去','要','还','很','啦','怎么','却'])
    for word in stopwords:
        if word in wordCount:
            del wordCount[word]
```

```python
wordCount
```

```python
keywords = sorted(wordCount.items(), key=lambda d:d[1], reverse = True)
```

```python
keywords
```

```python
import csv
with open ('keywords.csv','w', encoding = 'UTF-8') as k:
    writer = csv.writer(k)
    header = ['keyword','frequency']
    writer.writerow(header)
    writer.writerows(keywords)
```
