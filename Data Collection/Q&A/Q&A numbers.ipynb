```python
import csv
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import random
```

```python
headers = {
        'Authority':'www.mafengwo.cn',
        'Accept': '*/*',
        'Accept-Encoding': 'gzip, deflate,br',
        'Accept-Language': 'zh-CN,zh;q=0.9',
        'Connection': 'keep-alive',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'
        }
```

```python
df = pd.read_csv('mafengwo_url.csv')
```

```python
url_list = list(df['Url'])
name_list = list(df['Name'])
ary_member = []
for i in range(0,len(name_list)):
    dic_member = {}
    dic_member['Name'] = name_list[i]
    dic_member['Url'] = url_list[i]
    ary_member.append(dic_member)
```

```python
def Crawl_members_of_the_Question(member):
    question_url = member['Url'][:22] + '/wenda' + member['Url'][22:-5] + '/question.html'
    try:
        res = requests.get(question_url, headers=headers)
        soup_bs = BeautifulSoup(res.text, 'lxml')
        num_question = soup_bs.find('span',class_='_j_total_item').text
        member['Question'] = int(num_question[:-1])

        urls_question = []
        list_item = soup_bs.find('div',class_='wd-list _j_qa_list')
        for item in list_item.find_all('div',class_='wd-title'):
            url = item.find('a').get('href')
            urls_question.append('http://www.mafengwo.cn' + url)
        time.sleep(1)
        for url in urls_question:
            Crawl_member_of_the_content(url)
            time.sleep(1)
    except:
        print("Error:" + member['Name'] + ":Failed to crawl the number of questions or link to questions!")
    time.sleep(random.randint(2,5)) 
```

```python
def Crawl_members_of_the_Answer(member):
    answer_url = member['Url'][:22] + '/wenda' + member['Url'][22:-5] + '/answer.html'
    try:
        res = requests.get(answer_url, headers=headers)
        soup_bs = BeautifulSoup(res.text, 'lxml')
        num_answer = soup_bs.find('span',class_='_j_total_item').text
        member['Answer'] = int(num_answer[:-1])

        urls_answer = []
        list_item = soup_bs.find('div',class_='wd-list _j_qa_list')
        for item in list_item.find_all('div',class_='wd-title'):
            url = item.find('a').get('href')
            urls_answer.append('http://www.mafengwo.cn' + url)
        time.sleep(1)
        for url in urls_answer:
            Crawl_member_of_the_content(url)
            time.sleep(1)
    except:
        print("Error:" + member['Name'] + ":Failed to grab the number of answers or link to the question!")
    time.sleep(random.randint(2,5)) 
```

```python
def Crawl_member_of_the_content(content_url):
    try:
        #çˆ¬å–æé—®é—®é¢˜
        res = requests.get(content_url, headers=headers)
        soup_bs = BeautifulSoup(res.text, 'lxml')
        title_q = soup_bs.find('div',class_='q-title')
        title = title_q.find('h1').text.strip()
        content = soup_bs.find('div',class_='q-desc').text
        content_t = title + '\n'+ content
        #print(content_t)
        #çˆ¬å–å›ç­”
        for answer in soup_bs.find_all('div',class_='_j_long_answer_item'):
            answer_content = answer.find('div',class_='_j_answer_html')
            content_t = content_t + '\n' + answer_content.text
        with open('content.txt','a',encoding='utf_8_sig') as f:
            f.writelines(content_t)
    except:
        print('Error:Crawling questions are error-tolerant!')
```

```python
for member in ary_member:
    Crawl_members_of_the_Question(member)
    Crawl_members_of_the_Answer(member)
```
